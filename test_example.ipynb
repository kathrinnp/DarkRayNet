```python
import numpy as np
from sim_tool import antiproton_sim, primary_sim
sim = antiproton_sim()
sim_p = primary_sim()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['pgf.rcfonts'] = False
plt.rcParams['font.family'] = "DejaVu Sans"
plt.rcParams['font.serif'] = []
plt.rcParams['text.usetex'] = True
plt.rcParams['axes.formatter.useoffset'] = False
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['errorbar.capsize'] = 2
plt.rcParams['grid.linewidth'] = 2
plt.rcParams['axes.labelsize'] = 14

c_RWTH = {'b': '#00549F',  # blue
          'lb': '#8EBAE5', # light blue
          'dr': '#A11035', # dark red
          'g': '#57AB27',  # green
          't': '#006165',  # teal / petrol
          'o': '#F6A800',  # orange
          'lg': '#BDCD00', # light green
          'gr': '#646567', # gray
          'v': '#612158',  # violett
          'r': '#CC071E',  # red
          'tq': '#0098A1', # turquoise
          'p': '#7A6FAC'}  # purple
```

```python
test_mDM = 5 #log10(mDM/MeV)
test_mDM_N =np.array([4, 5, 6])

test_fs = np.ones(8)* 1/8 #equal branching fractions
test_fs_N = np.repeat([test_fs], 3, axis = 0) 

test_pp = np.array([1.70694e+00, 1.67743e+00, 2.44673e+00, 2.41569e+00, 7.56688e+03, 4.22478e-01, 2.79565e+28, 3.78306e-01, 1.75379e+01, 4.82479e-01, 3.05006e+00]) # arbitrary propagation model
test_pp_N = np.repeat([test_pp], 3, axis = 0)
```

```python
tot, DM, s, E = sim.single_sim(test_pp, test_mDM, test_fs)

plt.figure(figsize=(8,6))
plt.title('Cosmic Ray Antiproton Flux')
plt.plot(E, E**2.7 * tot, c = c_RWTH['b'], linestyle = 'dashed', label = 'total antiproton flux')
plt.plot(E, E**2.7 * s, c = c_RWTH['g'], linestyle = 'dotted', label = 'secondary antiproton flux')
plt.plot(E, E**2.7 * DM, c = c_RWTH['dr'], linestyle = 'dashdot', label = 'DM antiproton flux')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.ylabel('$\Phi \cdot E^{2.7}$ [GeV$^{1.7}$ m$^{-2}$ sr$^{-1}$ s$^{-1}$]')
plt.xlabel('E [GeV]')
plt.show()
```

```python
tot, DM, s, E = sim.N_sim(test_pp_N, test_mDM_N, test_fs_N)

plt.figure(figsize=(8,6))
plt.title('DM Component')
plt.plot(E, E**2.7 * DM[0], c = c_RWTH['b'], linestyle = 'dashed', label = r'$m_\mathrm{DM} = 10$ GeV')
plt.plot(E, E**2.7 * DM[1], c = c_RWTH['g'], linestyle = 'dotted', label = r'$m_\mathrm{DM} = 100$ GeV')
plt.plot(E, E**2.7 * DM[2], c = c_RWTH['dr'], linestyle = 'dashdot', label = r'$m_\mathrm{DM} = 1$ TeV')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.ylabel('$\Phi \cdot E^{2.7}$ [GeV$^{1.7}$ m$^{-2}$ sr$^{-1}$ s$^{-1}$]')
plt.xlabel('E [GeV]')
plt.show()
```

```python
p, D, He4, He3, E = sim_p.N_sim(test_pp_N)

plt.figure(figsize=(8,6))
plt.title('Cosmic Ray Proton Flux')
plt.plot(E, E**2.7 * p, c = c_RWTH['b'], linestyle = 'dashed', label = r'Protons')
plt.plot(E, E**2.7 * D, c = c_RWTH['g'], linestyle = 'dotted', label = r'Deuterium')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.ylabel('$\Phi \cdot E^{2.7}$ [GeV$^{1.7}$ m$^{-2}$ sr$^{-1}$ s$^{-1}$]')
plt.xlabel('E [GeV]')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title('Cosmic Ray Helium Flux')
plt.plot(E, E**2.7 * He4, c = c_RWTH['b'], linestyle = 'dashed', label = r'Helium 4')
plt.plot(E, E**2.7 * He3, c = c_RWTH['g'], linestyle = 'dotted', label = r'Helium 3')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.ylabel('$\Phi \cdot E^{2.7}$ [GeV$^{1.7}$ m$^{-2}$ sr$^{-1}$ s$^{-1}$]')
plt.xlabel('E [GeV]')
plt.show()
```
