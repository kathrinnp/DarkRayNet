```python
import numpy as np
import matplotlib.pyplot as plt
from simulation_tool import sim_tool
sim = sim_tool()
```

```python
test_mDM = np.array([10, 100, 1000])

test_fs = np.ones((3, 8))* 1/8 #equal branching fractions

test_pp = np.array([1.8, 1.79, 2.405, 2.357, 7.92e+03, 0.37, 2.05e+28, 0.419, 8.84, 0.09, 2.60]) # arbitrary propagation model
test_pp = np.repeat([test_pp], 3, axis = 0)
```

```python
out = sim.predict(["DM Antiprotons", "Secondary Antiprotons"], test_pp[0], DM_mass= test_mDM[1], DM_fs = test_fs[1])

DM_flux, DM_E = out[0]
S_flux, S_E = out[1]

plt.figure(figsize=(8,6))
plt.title('Cosmic Ray Antiproton Flux')
plt.plot(DM_E, DM_E**2.7 * (DM_flux + S_flux), c = 'b', linestyle = 'dashed', label = 'total antiproton flux')
plt.plot(S_E, S_E**2.7 * S_flux, c = 'g', linestyle = 'dotted', label = 'secondary antiproton flux')
plt.plot(DM_E, DM_E**2.7 * DM_flux, c = 'r', linestyle = 'dashdot', label = 'DM antiproton flux')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.ylabel('$\Phi \cdot E^{2.7}$ [GeV$^{1.7}$ m$^{-2}$ sr$^{-1}$ s$^{-1}$]')
plt.xlabel('E [GeV]')
plt.show()
```

```python
out = sim.predict(["DM Antiprotons"], test_pp, DM_mass= test_mDM, DM_fs = test_fs)

DM_flux, DM_E = out[0]

plt.figure(figsize=(8,6))
plt.title('DM Component')
plt.plot(DM_E, DM_E**2.7 * DM_flux[0], c = 'b', linestyle = 'dashed', label = r'$m_\mathrm{DM} = 10$ GeV')
plt.plot(DM_E, DM_E**2.7 * DM_flux[1], c = 'g', linestyle = 'dotted', label = r'$m_\mathrm{DM} = 100$ GeV')
plt.plot(DM_E, DM_E**2.7 * DM_flux[2], c = 'r', linestyle = 'dashdot', label = r'$m_\mathrm{DM} = 1$ TeV')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.ylabel('$\Phi \cdot E^{2.7}$ [GeV$^{1.7}$ m$^{-2}$ sr$^{-1}$ s$^{-1}$]')
plt.xlabel('E [GeV]')
plt.show()
```

```python
out = sim.predict(["Protons", "Deuterium", "Helium 3", "Helium 4"], test_pp, DM_mass= test_mDM, DM_fs = test_fs)

p, E = out[0]
D, E = out[1]
He3, E = out[2]
He4, E = out[3]

plt.figure(figsize=(8,6))
plt.title('Cosmic Ray Proton Flux')
plt.plot(E, E**2.7 * p[0], c = 'b', linestyle = 'dashed', label = r'Protons')
plt.plot(E, E**2.7 * D[0], c = 'g', linestyle = 'dotted', label = r'Deuterium')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.ylabel('$\Phi \cdot E^{2.7}$ [GeV$^{1.7}$ m$^{-2}$ sr$^{-1}$ s$^{-1}$]')
plt.xlabel('E [GeV]')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title('Cosmic Ray Helium Flux')
plt.plot(E, E**2.7 * He4[0], c = 'b', linestyle = 'dashed', label = r'Helium 4')
plt.plot(E, E**2.7 * He3[0], c = 'g', linestyle = 'dotted', label = r'Helium 3')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.ylabel('$\Phi \cdot E^{2.7}$ [GeV$^{1.7}$ m$^{-2}$ sr$^{-1}$ s$^{-1}$]')
plt.xlabel('E [GeV]')
plt.show()
```
